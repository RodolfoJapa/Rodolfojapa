```python
# Visualización 2
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
plt.plot(df_alertas["dia"], df_alertas["kurtosis"], marker="o", label="Kurtosis diaria")

plt.axhline(
    y=3,
    linestyle="--",
    label="Umbral de anomalía"
)

plt.title("Detección de Anomalías en Transacciones usando Kurtosis")
plt.xlabel("Día")
plt.ylabel("Kurtosis")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Visualización 3
anomalias = df_alertas[df_alertas["anomalia"]]

plt.figure(figsize=(12,5))
plt.plot(df_alertas["dia"], df_alertas["kurtosis"], label="Kurtosis")
plt.scatter(
    anomalias["dia"],
    anomalias["kurtosis"],
    s=80,
    label="Días anómalos"
)

plt.axhline(3, linestyle="--", label="Umbral")
plt.title("Días con Comportamiento Anómalo Detectados")
plt.xlabel("Día")
plt.ylabel("Kurtosis")
plt.legend()
plt.grid(True)
plt.show()
```
