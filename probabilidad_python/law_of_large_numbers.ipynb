```python
# ==============================
# Environment Setup
# ==============================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Global configuration
np.random.seed(42)
plt.rcParams["figure.figsize"] = (10, 6)
plt.style.use("default")
```

```python
# ==============================
# Data Generation
# ==============================

def generate_samples(
    n_samples: int,
    distribution: str = "uniform"
) -> np.ndarray:
    """
    Generate random samples from a given distribution.

    Parameters:
    - n_samples: number of observations
    - distribution: distribution type

    Returns:
    - numpy array of samples
    """

    if distribution == "uniform":
        return np.random.uniform(0, 1, n_samples)

    elif distribution == "normal":
        return np.random.normal(0, 1, n_samples)

    else:
        raise ValueError("Unsupported distribution")
```

```python
# ==============================
# LLN Core Logic
# ==============================

def compute_running_mean(samples: np.ndarray) -> np.ndarray:
    """
    Compute running (cumulative) mean.
    """
    return np.cumsum(samples) / np.arange(1, len(samples) + 1)
```

```python
# ==============================
# Simulation Pipeline
# ==============================

def lln_simulation(
    n_samples: int = 10000,
    distribution: str = "uniform"
):
    samples = generate_samples(n_samples, distribution)
    running_mean = compute_running_mean(samples)

    expected_value = 0.5 if distribution == "uniform" else 0.0

    return samples, running_mean, expected_value
```

```python
# ==============================
# Visualization
# ==============================

def plot_lln(running_mean, expected_value):
    plt.plot(running_mean, label="Sample Mean")
    plt.axhline(expected_value, color="red", linestyle="--", label="Expected Value")

    plt.title("Law of Large Numbers Demonstration")
    plt.xlabel("Number of Observations")
    plt.ylabel("Mean Value")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# ==============================
# Execution
# ==============================

samples, running_mean, expected_value = lln_simulation(
    n_samples=20000,
    distribution="uniform"
)

plot_lln(running_mean, expected_value)
```
